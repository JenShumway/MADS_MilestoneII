```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wbd = pd.read_csv('../WorldBankDatasets/Cleaned/AllMerged_Threshold_85_n3_MMRTNE.csv')[['Country', 'Year', 'SH.STA.MMRT.NE']]
oecd = pd.read_csv('../OECD/Cleaned/HEALTH_MERGED_Threshold_80_n3.csv')[['Country', 'Year', 'MATIMATM']]
```

```python
df = pd.merge(wbd, oecd, how='inner', on=['Year', 'Country'])
df
```

```python
oecd['Year'].describe(), wbd['Year'].describe()
```

```python
# Extract the 'Year' and Maternal Mortality Rate columns for comparison
df = df[df['Country'] == 'TUR']
years = df['Year']
column1 = df['SH.STA.MMRT.NE']
column2 = df['MATIMATM']


# Create a line plot for each column
plt.figure(figsize=(10, 6))
plt.plot(years, column1, label='SH.STA.MMRT.NE', marker='o', linestyle='-', color='b')
plt.plot(years, column2, label='MATIMATM', marker='o', linestyle='-', color='g')


# Add labels
plt.title('Comparison of Maternal Mortality Indicators (Turkey)')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
